```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Sex': ['F', 'M', 'M', 'M']
}
df = pd.DataFrame(data)
print(df)
```

```python
# Select single column
print(df['Name'])
# Select multiple columns
print(df[['Name', 'Age']])
# Select row by index
print(df.iloc[0])   # first row by position
print(df.loc[0])    # first row by label
# Select subset of rows and columns
print(df.loc[0:1, ['Name', 'Sex']])
```

```python
# Filter rows: Age > 28
print(df[df['Age'] > 28])
# Sort by Age (descending)
print(df.sort_values('Age', ascending=False))
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Add new column (Age after 5 years)
df['Age_in_5yrs'] = df['Age'] + 5
print("\nAfter adding column:\n", df)

# Drop the new column
df = df.drop('Age_in_5yrs', axis=1)  # axis=1 means column
print("\nAfter dropping column:\n", df)
```

```python
# New dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob'],
    'Score': [85, 90, 95, 80, 70]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Group by Name and calculate mean
grouped = df.groupby('Name').mean(numeric_only=True)   # mean for numeric columns
print("\nGrouped by Name (mean score):\n", grouped)
```

```python
data = {'Name':['Alice','Bob','Charlie','David'],
        'Age':[25, None, 35, 40]}
df = pd.DataFrame(data)
 
# Fill missing value
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)
 
# Drop rows with missing values
df = df.dropna()
```
